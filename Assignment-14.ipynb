## RGBA color values are an extension of RGB color values with an alpha channel - which specifies the opacity for a color. An RGBA color value is specified with: rgba(red, green, blue, alpha).

```python
'''This function takes a color name string as its first argument, and the string 'RGBA' as its second argument, 
and it returns an RGBA tuple.'''
from PIL import ImageColor
ImageColor.getcolor('CornflowerBlue', 'RGBA')
```

## A box tuple is a tuple value of four integers: the left edge x-coordinate, the top edge y-coordinate, the width, and the height, respectively.

```python
from PIL import Image
vm=Image.open('Venom.png')
vm
```

```python
## Size function returns width and height of the image in tuple
vm.size
```

```python
## I will use this crop function to crop that image
croppedIm = vm.crop((0, 50, 50, 50))
## I will use save function to save the file with new name
croppedIm.save('cropped.png')
```

## The ImageDraw module contains code to draw on images

```python
## ImageDraw object has methods like rectangle, line, Point. They are returned by passing image obj to draw ()
from PIL import Image, ImageDraw
## creating white image
createimage = Image.new('RGBA', (200, 200), 'white')
## usingImageDraw function i can draw any shape on image
draw = ImageDraw.Draw(createimage)
draw.rectangle((20, 30, 60, 60), fill='blue')
createimage.save('PaintedImage.png')
Image.open('PaintedImage.png')
```
